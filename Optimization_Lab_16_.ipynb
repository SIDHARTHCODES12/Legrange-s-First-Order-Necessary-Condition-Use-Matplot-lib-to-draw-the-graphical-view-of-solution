```python
#RegNo 2367413    Name:Nishant    Prog:Legrange’s FONC graphical
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse, Circle
center = (0, 0)
width = 2
height = 2*(1/(2**.5))
fig, ax = plt.subplots(figsize=(6, 6))
ellipse = Ellipse(center, width, height, edgecolor='b', facecolor='none'
                  , linewidth=1)
ax.add_patch(ellipse)
circle1 = Circle(center,width/2 , edgecolor='g', facecolor='none'
                 , linewidth=1)
ax.add_patch(circle1)
circle2 = Circle(center, height/2, edgecolor='r', facecolor='none'
                 , linewidth=1)
ax.add_patch(circle2)
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_title("visualization of Lagrange's first order necessary condition")
plt.scatter([-1*(width/2),0,(width/2),0,0],[0,0,0,height/2,-1*height/2],color='red'
            , marker='o', s=10,zorder=10)
plt.text(-1*(width/2),0,'[-1,0]',color='r', fontsize=8, ha='right', va='top')
plt.text(1*(width/2),0,'[1,0]',color='r', fontsize=8, ha='right', va='top')
plt.text(0,-1*height/2,r'$[0,-1/\sqrt{2}]$',color='r', fontsize=8, ha='right', va='top')
plt.text(0,1*height/2,r'$[0,1/\sqrt{2}]$',color='r', fontsize=8, ha='right', va='bottom')
plt.grid(True)
plt.show()
```
