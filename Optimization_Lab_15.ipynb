```python
#RegNo 2367413    Name:Nishant    Prog:Optimal solution using matplotlib(lab15)
import numpy as np
import matplotlib.pyplot as plt
x1_values = np.linspace(-1, 5, 400)
x2_values = np.linspace(-3, 1, 400)
X1, X2 = np.meshgrid(x1_values, x2_values)
H = (X1-2)**2
G = (X2+1)**3
plt.figure(figsize=(8, 6))
contour_h = plt.contour(X1, X2, H, colors='r')
contour_g = plt.contour(X1, X2, G,levels=[0], colors='g')
plt.fill_between(x1_values, -3, -1, color='blue', alpha=0.3,
                 label='$g(x_1, x_2) = (x_2 + 1)^3 \leq 0$')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Graphical Method - Feasible Region')
plt.legend()
plt.clabel(contour_h, inline=True, fontsize=8)
plt.grid(True)
plt.show()
```
